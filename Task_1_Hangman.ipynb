```python
import random

# Hangman shapes for each incorrect guess
HANGMAN_SHAPES = [
    """
      ------
      |    |
           |
           |
           |
           |
    =========""",
    """
      ------
      |    |
      O   |
            |
            |
            |
    =========""",
    """
      ------
      |    |
      O   |
      |    |
            |
            |
    =========""",
    """
      ------
      |    |
      O   |
     /|   |
            |
            |
    =========""",
    """
      ------
      |     |
      O    |
     /|\   |
             |
             |
    =========""",
    """
      ------
      |     |
      O    |
     /|\   |
     /       |
              |
    =========""",
    """
      ------
      |    |
      O   |
     /|\  |
     / \   |
            |
    ========="""
]

# List of words for the hangman game
WORDS = ["apple", "banana", "pomegranate", "grapes", "kiwi"]

def random_word():
    return random.choice(WORDS)

def display_hangman(incorrect_guesses):
    print(HANGMAN_SHAPES[incorrect_guesses])

def play_game():
    word = random_word()
    guessed_word = ["_"] * len(word)
    guessed_letters = set()
    incorrect_guesses = 0

    print("Welcome to Hangman!")
    while incorrect_guesses < len(HANGMAN_SHAPES) - 1 and "_" in guessed_word:
        print("\nCurrent word: " + " ".join(guessed_word))
        display_hangman(incorrect_guesses)
        guess = input("Guess a letter: ").lower()

        if guess in guessed_letters:
            print("You already guessed that letter.")
        elif guess in word:
            guessed_letters.add(guess)
            for idx, letter in enumerate(word):
                if letter == guess:
                    guessed_word[idx] = letter
        else:
            guessed_letters.add(guess)
            incorrect_guesses += 1
            print(f"Wrong guess! You have {len(HANGMAN_SHAPES) - 1 - incorrect_guesses} tries left.")

    if "_" not in guessed_word:
        print("\nCongratulations! You've guessed the word: " + word)
    else:
        display_hangman(incorrect_guesses)
        print("\nSorry, you've been hanged! The word was: " + word)

if __name__ == "__main__":
    play_game()
```
